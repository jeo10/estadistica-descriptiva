```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('files/cars.csv')
```

```python
#desviación estandar
df.price_usd.std()
```

```python
#rango = max - min
rango = df.price_usd.max() - df.price_usd.min()
rango
```

```python
#Quartiles
median = df.price_usd.median()
Q1 = df.price_usd.quantile(q=0.25)
Q3 = df.price_usd.quantile(q=0.75)
```

```python
#rango intercuartil
iqr = Q3 - Q1
iqr
```

```python
#Límites para detección de Outliers (datos simétricamente distribuidos)
minlimit = Q1 - 1.5*iqr
maxlimit = Q3 + 1.5*iqr
print(minlimit, maxlimit)
#resultado porque los datos no están simétricamente distribuidos
```
